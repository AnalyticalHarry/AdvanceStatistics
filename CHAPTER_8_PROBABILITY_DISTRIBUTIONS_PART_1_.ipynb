# **Probability Distributions**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson
from scipy.stats import norm, chi2, t, lognorm, expon
```

```python
# probability of sucess for Bernoulli and Binomial distributions
p = 0.3
# number of trails for binomial distribution
n = 10
# average rate for poisson distribution
lambda_ = 3

# values for x - axis
x = np.arange(0,11)
```

```python
# probability mass function for Bernoulli distribution
pmf_bernoulli = bernoulli.pmf(x, p)
# bernoulli Distribution
plt.figure(figsize=(5, 5))
plt.bar(x, pmf_bernoulli, align='center', color='black')
plt.title('Bernoulli Distribution (p=0.3)')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(True, ls='--', alpha=0.2)
plt.show()
```

```python
# probability mass function for Binomial distribution
pmf_binomial = binom.pmf(x, n, p)
# binomial Distribution
plt.figure(figsize=(5, 5))
plt.bar(x, pmf_binomial, align='center', color='black')
plt.title('Binomial Distribution (n=10, p=0.3)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(True, ls='--', alpha=0.2)
plt.show()
```

```python
# probability mass function for Poisson distribution
pmf_poisson = poisson.pmf(x, lambda_)
# poisson Distribution
plt.figure(figsize=(5, 5))
plt.bar(x, pmf_poisson, align='center', color='black')
plt.title('Poisson Distribution (λ=3)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.grid(True, ls='--', alpha=0.2)
plt.show()
```

```python
# parameters for the distributions
mean = 0
std_dev = 1

# degrees of freedom for Chi-square and Student-t distributions
df = 2

# shape parameter for Log-Normal distribution
sigma_lognorm = 0.5

# rate parameter for Exponential distribution
lambda_exp = 0.5

# values for x-axis
x = np.linspace(-5, 5, 1000)

# probability density function for Normal distribution
normal_pdf = norm.pdf(x, mean, std_dev)

# probability density function for Chi-square distribution
chi2_pdf = chi2.pdf(x, df)

# probability density function for Student-t distribution
student_t_pdf = t.pdf(x, df)

# probability density function for Log-Normal distribution
lognorm_pdf = lognorm.pdf(x, sigma_lognorm)

# probability density function for Exponential distribution
exponential_pdf = expon.pdf(x, scale=1/lambda_exp)
```

```python
plt.figure(figsize=(12, 8))

# normal Distribution
plt.subplot(2, 3, 1)
plt.plot(x, normal_pdf, label='Normal Distribution', color='black')
plt.title('Normal Distribution (μ=0, σ=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# chi-square Distribution
plt.subplot(2, 3, 2)
plt.plot(x, chi2_pdf, label='Chi-square Distribution', color='black')
plt.title('Chi-square Distribution (df=2)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# student-t Distribution
plt.subplot(2, 3, 3)
plt.plot(x, student_t_pdf, label='Student-t Distribution', color='black')
plt.title('Student-t Distribution (df=2)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# log-Normal Distribution
plt.subplot(2, 3, 4)
plt.plot(x, lognorm_pdf, label='Log-Normal Distribution', color='black')
plt.title('Log-Normal Distribution (σ=0.5)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# exponential Distribution
plt.subplot(2, 3, 5)
plt.plot(x, exponential_pdf, label='Exponential Distribution', color='black')
plt.title('Exponential Distribution (λ=0.5)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

plt.tight_layout()
plt.show()
```
