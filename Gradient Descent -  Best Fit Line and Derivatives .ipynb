## GRADIENT DESCENT

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
```

### HEMANT THAPA

#### Example 1

```python
data = [
    (1, 2),
    (2, 4),
    (3, 5),
    (4, 4)
]
```

```python
m = 0.5
b = 0.5
learning_rate = 0.01
iterations = 50
```

```python
iteration_list = []
mse_list = []
y_pred_list = []
dm_list = []
db_list = []
m_new_list = []
b_new_list = []
```

```python
for iteration in range(iterations):
    # Calculate predicted values
    y_pred = [m * xi + b for xi, _ in data]
    
    # Calculate MSE
    mse = sum([(yi - ypi)**2 for (_, yi), ypi in zip(data, y_pred)]) / len(data)
    
    # Calculate derivatives
    dm = (-2/len(data)) * sum([xi * (yi - ypi) for (xi, yi), ypi in zip(data, y_pred)])
    db = (-2/len(data)) * sum([yi - ypi for (_, yi), ypi in zip(data, y_pred)])
    
    # Update m and b
    m_new = m - learning_rate * dm
    b_new = b - learning_rate * db
    
    # Append values to lists
    iteration_list.append(iteration)
    mse_list.append(mse)
    y_pred_list.append(y_pred)
    dm_list.append(dm)
    db_list.append(db)
    m_new_list.append(m_new)
    b_new_list.append(b_new)

    # Print iteration details
    print(f"Iteration {iteration + 1}: MSE = {mse:.4f}, m = {m:.4f}, b = {b:.4f}, dm = {dm:.4f}, db = {db:.4f}")
    
    # Update m and b for the next iteration
    m = m_new
    b = b_new
```

```python
plt.figure(figsize=(12, 7))

# Plot Mean Squared Error (MSE) curve
plt.subplot(2, 2, 1)
plt.plot(iteration_list, mse_list, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Gradient Descent for Linear Regression (MSE Curve)')
plt.grid(True)

# Plot m and b values
plt.subplot(2, 2, 2)
plt.plot(iteration_list, m_new_list, label='m')
plt.plot(iteration_list, b_new_list, label='b')
plt.xlabel('Iteration')
plt.ylabel('Parameter Value')
plt.title('Parameter Update During Gradient Descent')
plt.legend()
plt.grid(True)

# Plot dm and db values
plt.subplot(2, 2, 3)
plt.plot(iteration_list, dm_list, label='dm')
plt.plot(iteration_list, db_list, label='db')
plt.xlabel('Iteration')
plt.ylabel('Derivative Value')
plt.title('Derivative Update During Gradient Descent')
plt.legend()
plt.grid(True)

# Plot predicted y values and data points
plt.subplot(2, 2, 4)
for i, y_pred in enumerate(y_pred_list):
    plt.plot([xi for xi, _ in data], y_pred, label=f'Iteration {i+1}')
for xi, yi in data:
    plt.scatter(xi, yi, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Predicted y Values During Gradient Descent')
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### Example 3

```python
data = [
    (1, 2),
    (2, 4),
    (3, 5),
    (4, 4)
]
```

```python
m = 0.5
b = 0.5
learning_rate = 0.01
iterations = 15
```

```python
iteration_list = []
mse_list = []
y_pred_list = []  
```

```python
for iteration in range(iterations):
    y_pred_iteration = []  # List to store predicted values for this iteration
    
    for xi, yi in data:
        # Calculate predicted value
        y_pred = m * xi + b
        y_pred_iteration.append(y_pred)
    
    y_pred_list.append(y_pred_iteration)
    
    # Calculate MSE
    mse = sum([(yi - ypi)**2 for yi, ypi in zip([y for _, y in data], y_pred_iteration)]) / len(data)
    
    # Calculate derivatives
    dm = (-2/len(data)) * sum([xi * (yi - ypi) for (xi, yi), ypi in zip(data, y_pred_iteration)])
    db = (-2/len(data)) * sum([yi - ypi for (_, yi), ypi in zip(data, y_pred_iteration)])
    
    # Update m and b
    if iteration > 0:  # Start updating from the second iteration
        m -= learning_rate * dm
        b -= learning_rate * db
    
    # Append values to lists
    iteration_list.append(iteration)
    mse_list.append(mse)
    
    # Print iteration details
    if iteration == 0:
        print(f"Iteration {iteration + 1} - Initial values: m = {m:.3f}, b = {b:.3f}")
    else:
        print(f"Iteration {iteration + 1} - Updated values: m = {m:.3f}, b = {b:.3f}")
    print("Calculate predicted values y_pred:")
    
    for (xi, _), y_pred in zip(data, y_pred_iteration):
        y_pred_calc = m * xi + b  # Correct calculation for predicted value
        print(f"For ({xi}, {_}): y_pred = {m:.3f} * {xi} + {b:.3f} = {y_pred_calc:.3f}")
```
