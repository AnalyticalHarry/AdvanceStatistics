### 1. MEAN

Definition: The mean is calculated as the sum of all values in a dataset divided by the total number of values.

Calculation: Mean = (Sum of all values) / (Total number of values)

Interpretation: The mean represents the average value of the dataset. It is sensitive to extreme values or outliers, as it takes into account all values in the dataset.

The population mean and the sample mean are statistical measures that provide an estimate of the central tendency or average of a dataset. However, they differ in the way they are calculated and the population they represent.

#### 1. Population Mean:

The population mean, denoted by the symbol  (mu), represents the average value of a variable in an entire population.
It is calculated by summing up all the values in the population and dividing the sum by the total number of values in the population.

The formula for the population mean is:  = (X) / N, where X represents the sum of all values in the population, and N represents the total number of values in the population.

The population mean provides a measure of the central tendency for the entire population.

#### 2. Sample Mean:

The sample mean, denoted by the symbol x (x-bar), represents the average value of a variable in a sample selected from a larger population.

It is calculated by summing up all the values in the sample and dividing the sum by the total number of values in the sample.

The formula for the sample mean is: x = (X) / n, where X represents the sum of all values in the sample, and n represents the total number of values in the sample.

The sample mean provides an estimate of the population mean based on the observed data in the sample.

It is often used as a representative value to make inferences about the population mean.

#### 3. Key Differences:

The population mean is calculated using the entire population, while the sample mean is calculated using a subset of the population (the sample).

The population mean represents the true average of the entire population, whereas the sample mean estimates the population mean based on the observed data in the sample.

The sample mean is more susceptible to sampling variability and may differ from the population mean.

The sample mean is often used to make inferences and draw conclusions about the population mean.

#### Suppose we want to determine the average height of all adults in a particular country. We have the following data:

Population: [170 cm, 165 cm, 175 cm, 180 cm, 160 cm]

Sample: [170 cm, 165 cm, 175 cm]

To calculate the population mean:

#### 1. Population Mean ():

Sum all the values in the population: 170 cm + 165 cm + 175 cm + 180 cm + 160 cm = 850 cm

Divide the sum by the total number of values in the population (5 in this case):  = 850 cm / 5 = 170 cm

The population mean is 170 cm.

To calculate the sample mean:

#### 2. Sample Mean (x):

Sum all the values in the sample: 170 cm + 165 cm + 175 cm = 510 cm

Divide the sum by the total number of values in the sample (3 in this case): x = 510 cm / 3 = 170 cm

The sample mean is also 170 cm.

### 2. MEDIAN

Definition: Median is defined as the middle number in a given data set if the given numbers or observations are arranged in either ascending or descending patterns. Hence Median refers to the value of the middlemost observation present in the given data set after organizing the data in ascending or descending order. 

Calculation: If the dataset has an odd number of values, the median is the middle value. If the dataset has an even number of values, the median is the average of the two middle values.

Interpretation: The median represents the value that divides the dataset into two equal halves. It is less affected by extreme values or outliers, making it a robust measure of central tendency.

Lower limit (l): The lower limit of the median class is the smallest value within that class. For the class 20-30, the lower limit is 20.

Frequency (f): The frequency of the median class represents the number of observations within that class. In this case, the frequency of the class 20-30 is 12, indicating that there are 12 students within the range of 20-30.

### 3. Mode

Definition: Mode is defined as the value that refers to the most frequently occurring number or observation in a given data set. In simple words, the number/ observations having the highest frequency is known as the mode of given data. The range here denotes the difference between the highest and lowest values of observations in the given data set. It is to be noted here that MODE stands for Most Often Data Entered. It is also remembered that the data set may have no mode, 1 mode, or more than 1 mode. On the basis of the number of modes of given data, the data set can be known as unimodal, bimodal, trimodal, or multimodal.

Calculation: The mode is determined by identifying the value(s) with the highest frequency.

Interpretation: The mode represents the most common value(s) in the dataset. Unlike the mean and median, the mode can be used with both numerical and categorical data.

#### Where:

L = Lower limit Mode of modal class

fm = Frequency of modal class

f1 = Frequency of class preceding the modal class

f2= Frequency of class succeeding the modal class

h = Size of class interval

1. Unimodal: A unimodal distribution has a single peak or mode. An example could be the distribution of heights of adult males, where most of the observations cluster around a single height value.


2. Bimodal: A bimodal distribution has two distinct peaks or modes. An example could be the distribution of test scores in a classroom, where there is a group of students who perform well and another group who perform poorly, resulting in two peaks in the distribution.


3. Trimodal: A trimodal distribution has three distinct peaks or modes. An example could be the distribution of daily temperature readings in a city during different seasons. The distribution may show peaks in the temperature values corresponding to winter, spring, and summer.


4. Multimodal: A multimodal distribution has more than three distinct peaks or modes. An example could be the distribution of income levels in a country, where there might be peaks representing different income groups such as low-income, middle-income, and high-income.

```python
def MeanValue(random_list):
    total_sum = 0

    for i in range(len(random_list)):
        total_sum += random_list[i]
        list_length = len(random_list)
    
    print("Sum of all elements:", total_sum)
    print("Length of the list:", list_length)
    print("Mean:", round(total_sum / list_length, 2))
```

```python
def MedianValue(dataset):
    sorted_dataset = sorted(dataset)
    n = len(sorted_dataset)

    if n % 2 == 1:
        median = sorted_dataset[(n - 1) // 2]
    else:
        middle_right = n // 2
        middle_left = middle_right - 1
        median = (sorted_dataset[middle_left] + sorted_dataset[middle_right]) / 2

    print("Median:", median)
```

```python
MedianValue(random_list)
```

```python
st.mode(random_list)
```

```python
def ModeValue(dataset):
    frequency_dict = {}
    max_freq = 0
    modes = []

    for value in dataset:
        if value in frequency_dict:
            frequency_dict[value] += 1
        else:
            frequency_dict[value] = 1

        if frequency_dict[value] > max_freq:
            max_freq = frequency_dict[value]

    for value, frequency in frequency_dict.items():
        if frequency == max_freq:
            modes.append(value)

    print("Mode(s):", modes)
```

```python
dataset = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
ModeValue(dataset)
```

```python
from statistics import median

def calculate_mode(data):
    frequency_dict = {}
    max_freq = 0
    modes = []

    for value in data:
        if value in frequency_dict:
            frequency_dict[value] += 1
        else:
            frequency_dict[value] = 1

        if frequency_dict[value] > max_freq:
            max_freq = frequency_dict[value]

    for value, frequency in frequency_dict.items():
        if frequency == max_freq:
            modes.append(value)

    return modes
```

```python
study_hours = [3, 2, 4, 1, 2, 3, 2, 4, 3, 2, 1, 3, 2, 4, 3]
exam_scores = [70, 65, 80, 60, 65, 75, 70, 85, 80, 75, 65, 75, 70, 85, 80]
exercise_hours = [1, 0, 2, 0, 1, 1, 0, 2, 1, 0, 0, 1, 1, 2, 1]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# Plot Study Hours
axs[0].hist(study_hours, bins=range(max(study_hours) + 2), edgecolor='black')
axs[0].set_xlabel('Study Hours')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Study Hours')

mean_study = sum(study_hours) / len(study_hours)
axs[0].axvline(mean_study, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_study:.2f}')

mode_study = calculate_mode(study_hours)
axs[0].axvline(mode_study[0], color='green', linestyle='dashed', linewidth=2, label=f'Mode: {mode_study[0]}')

median_study = median(study_hours)
axs[0].axvline(median_study, color='purple', linestyle='dashed', linewidth=2, label=f'Median: {median_study}')

axs[0].legend()

# Plot Exam Scores
axs[1].hist(exam_scores, bins=range(min(exam_scores), max(exam_scores) + 5, 5), edgecolor='black')
axs[1].set_xlabel('Exam Scores')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Exam Scores')

mean_scores = sum(exam_scores) / len(exam_scores)
axs[1].axvline(mean_scores, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_scores:.2f}')

mode_scores = calculate_mode(exam_scores)
axs[1].axvline(mode_scores[0], color='green', linestyle='dashed', linewidth=2, label=f'Mode: {mode_scores[0]}')

median_scores = median(exam_scores)
axs[1].axvline(median_scores, color='purple', linestyle='dashed', linewidth=2, label=f'Median: {median_scores}')

axs[1].legend()

# Plot Exercise Hours
axs[2].hist(exercise_hours, bins=range(max(exercise_hours) + 2), edgecolor='black')
axs[2].set_xlabel('Exercise Hours')
axs[2].set_ylabel('Frequency')
axs[2].set_title('Exercise Hours')

mean_exercise = sum(exercise_hours) / len(exercise_hours)
axs[2].axvline(mean_exercise, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_exercise:.2f}')

mode_exercise = calculate_mode(exercise_hours)
axs[2].axvline(mode_exercise[0], color='green', linestyle='dashed', linewidth=2, label=f'Mode: {mode_exercise[0]}')

median_exercise = median(exercise_hours)
axs[2].axvline(median_exercise, color='purple', linestyle='dashed', linewidth=2, label=f'Median: {median_exercise}')

axs[2].legend()

plt.tight_layout()
plt.show()
```
